# Chebyshev's Inequality

### For a random variable $X$ with mean $\mu$ and finite variance $\sigma^2 < \infty$,
### Chebyshev's inequality states:
### $$
 Pr(|X - \mu| \ge k \sigma) \le \frac{1}{k^2}.
 $$



## QUESTION 1: 
### Suppose $X \sim N(\mu,\sigma^2)$.
### Calculate $Pr(|X - \mu| \ge k \sigma)$ for a number of different values of $k$
### and compare the exact results with the bound given by Chebyshev's inequality.



## SOLUTION


$$
Pr(|X-\mu| \ge k \sigma)
= Pr( X-\mu \ge k \sigma) + Pr(X-\mu \le -k \sigma)
= Pr( X \ge \mu + k \sigma)+ Pr(X \le \mu -k \sigma).
$$
But for the normal distribution we evaluate cumulative probabilities
by standardising.
Thus
$$
Pr(|X-\mu| \ge k \sigma)
= Pr( \frac{X-\mu}{\sigma} \ge k) + Pr(\frac{X-\mu}{\sigma} \le -k)
= Pr(Z \ge k) + Pr(Z \le -k)
$$
where ${\displaystyle Z = \frac{X-\mu}{\sigma} \sim N(0,1)}$ 
with $\Phi(z)=Pr(Z \le z)$.  

Hence

$$Pr(|X-\mu| \ge k \sigma) = (1- Pr(Z \le k) + Pr(Z \le -k)\\
= (1-\Phi(k)) + \Phi(-k) \\
= (1-\Phi(k))+(1-\Phi(k))=2(1-\Phi(k)).$$


The probabilities can be generated for different $k$ using the R code below.
The probabilities and Chebyshev bound are shown in table 1

```r
k=c(0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0)    # Assign some k values.
prob=2*(1-pnorm(k))                     # Exact probabilities.
chebyshev=1/k^2                         # Chebyshev's bound.
```

```r
options(repr.plot.width = 8, repr.plot.height = 8)
```

```r
DF=data.frame(K=k,ExactProb=prob, ChebyBound=chebyshev)
```

```r
DF
###NOTE THAT for every value of k, the exact probability is lower then the corresponding Chebyshev bound !
```

```r
plot(k,prob,type="l",ylim=c(0,1))
points(k,chebyshev,type="l",lty=2,col='red')
```
