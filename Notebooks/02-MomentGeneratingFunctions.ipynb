### Plot the analytical result as well as the empirical estimates of E(e^(tx)) for the exponential distribution

```r
options(repr.plot.width = 8, repr.plot.height = 8)

#replicate
set.seed(110)
sims = 1000

#define a simple parameter
lambda = 5

#generate the r.v.
X = rexp(sims, lambda)

#define t (in an interval near 0)
t = seq(from = 0, to = 2, length.out = 10)

#calculate the empirical and analytical MGFs
MGF.a = sapply(t, function(t) lambda/(lambda  - t))
MGF.e = sapply(t, function(t) mean(exp(t*X)))

#plot the MGFs, should match
plot(t, MGF.e, main = "Exponential MGF", xlab = "t",
     ylab = "", type = "p", pch = 16, col = "black", lwd = 3)
lines(t, MGF.a, type = "l",
      lwd = 3, col = "red")



legend("topleft", legend = c("Analytical MGF", "Empirical MGF"),
       lty=c(1,1), lwd=c(2.5,2.5),
       col=c("red", "black"))
```

## Comparing the sum of two Pois(1) random variables to a Pois(2) random variable in R.
### Prove that: sum of Poisson random variables is Poisson, and even show that the parameter of this new Poisson is the sum of the parameters of the individual Poissons.


```r

#replicate
set.seed(110)
sims = 1000

#generate the r.v.'s
X = rpois(sims, 1)
Y = rpois(sims, 1)
Z = rpois(sims, 2)

#set a 1x2 grid
par(mfrow = c(1,2))

#graphics; histograms break should match
hist(X + Y, main = "X + Y", xlab = "",
     col = rgb(1, 0, 0, 1/4), breaks = 0:10)
hist(Z, main = "Z", xlab = "",
     col = rgb(0, 1, 0, 1/4), breaks = 0:10)
```
